```python
import numpy as np
```

```python
# Load the .npy file
predictions = np.load('real_traffic/predictions/predictions_delay_real_traces.npy')

# Check the shape and size of the array
print("Shape of the array:", predictions.shape)
print("Total number of elements:", predictions.size)
```

```python
# Calculate statistical measures
print("Mean:", np.mean(predictions))
print("Standard Deviation:", np.std(predictions))
print("Minimum:", np.min(predictions))
print("Maximum:", np.max(predictions))
```

```python
# Assuming 'mape_array' is your NumPy array of MAPEs
mean_mape = np.mean(predictions)
median_mape = np.median(predictions)
std_dev_mape = np.std(predictions)

print("Mean MAPE:", mean_mape)
print("Median MAPE:", median_mape)
print("Standard Deviation of MAPE:", std_dev_mape)
```
